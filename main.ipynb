```python
import pandas as pd
import numpy as np

col_names = np.array(["duration","protocol_type","service","flag","src_bytes",
    "dst_bytes","land","wrong_fragment","urgent","hot","num_failed_logins",
    "logged_in","num_compromised","root_shell","su_attempted","num_root",
    "num_file_creations","num_shells","num_access_files","num_outbound_cmds",
    "is_host_login","is_guest_login","count","srv_count","serror_rate",
    "srv_serror_rate","rerror_rate","srv_rerror_rate","same_srv_rate",
    "diff_srv_rate","srv_diff_host_rate","dst_host_count","dst_host_srv_count",
    "dst_host_same_srv_rate","dst_host_diff_srv_rate","dst_host_same_src_port_rate",
    "dst_host_srv_diff_host_rate","dst_host_serror_rate","dst_host_srv_serror_rate",
    "dst_host_rerror_rate","dst_host_srv_rerror_rate", "class", 'difficulty level'])
```

```python
data = pd.read_csv("./NSL-KDD/KDDTrain+.txt", names = col_names)
data.head(10)
```

```python
# train_df.info()
data.isnull().sum()
data.dropna(inplace=True, axis=1)
```

```python
data = data.drop(columns=['num_outbound_cmds', 'srv_count', 'dst_bytes', 'src_bytes', 
                                  'land', 'is_host_login', 'urgent', 'num_failed_logins', 'num_shells'])

data.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder


def encode_label(data, label):
    label_encoder = LabelEncoder()

    encode_label = label_encoder.fit_transform(data[label])
    data[label] = encode_label

encode_label(data, "protocol_type")
encode_label(data, "service")
encode_label(data, "flag")
# data.head(10)
```

```python
Dos = ['land','neptune','smurf','pod','back','teardrop']
Probe = ['portsweep','ipsweep','satan','nmap']
U2R = ['buffer_overflow','loadmodule','perl','rootkit']

def encode_attack(vec):
    if vec in Dos:
        return "Dos"
    elif vec in Probe:
        return "Probe"
    elif vec in U2R:
        return "U2R"
    elif vec == "normal":
        return "normal"
    else:
        return "R2L"
    
data["attack_type"] = data["class"].apply(encode_attack)

data.head(10)
```

```python
data = data.drop(columns=['class', 'difficulty level'])
data.head(10)
```

```python
from scipy.stats import zscore
def encode_zscore(data, name):
    _zscore = zscore(data[name])
    data[name] = _zscore

cols = data.columns.values
for col in cols:
    count = data[col].value_counts()
    # print(count)
    for value, count in count.items():
        try:
            if int(value) < -1 or int(value) > 1:
                encode_zscore(data, col)
                break
        except ValueError:
            print()
        
data.head(10)
```

```python
from sklearn.model_selection import train_test_split
print(data.shape)
y = data[["attack_type"]]
X = data.drop(["attack_type", ], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
print("Data training: ", end="")
print(X_train.shape, y_train.shape)
print("Data test: ", end="")
print(X_test.shape, y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import time


clfd = DecisionTreeClassifier(criterion ="entropy", max_depth = 4)
start_time = time.time()
clfd.fit(X_train, y_train.values.ravel())
end_time = time.time()
print("Training time: ", end_time-start_time)



print("Train score is:", clfd.score(X_train, y_train))
print("Test score is:", clfd.score(X_test, y_test))
y_test_pred = clfd.predict(X_test)
X_test['attack-type'] = y_test_pred
print(X_test.head(10))

```
